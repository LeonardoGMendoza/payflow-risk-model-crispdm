## 🎯 3. Definição do Target (Variável-Alvo)

O **NPS (Net Promoter Score)** é nossa variável alvo. Seguindo a metodologia padrão de mercado, vamos classificar os clientes em:

| Classificação | Score NPS | Significado |
|---|---|---|
| 🟢 **Promotor** | 9–10 | Cliente fidelizado, recomenda a marca |
| 🟡 **Neutro** | 7–8 | Satisfeito, mas vulnerável à concorrência |
| 🔴 **Detrator** | 0–6 | Insatisfeito, pode gerar boca-a-boca negativo |

```python
# Criar coluna de classificação NPS
def classificar_nps(score):
    if score >= 9:   return 'Promotor'
    elif score >= 7: return 'Neutro'
    else:            return 'Detrator'

df['nps_categoria'] = df['nps_score'].apply(classificar_nps)
df['is_detrator']   = (df['nps_score'] < 7).astype(int)

contagem = df['nps_categoria'].value_counts()
print('📊 Distribuição das Categorias NPS:')
for cat, qtd in contagem.items():
    pct = qtd / len(df) * 100
    emoji = '🟢' if cat == 'Promotor' else ('🟡' if cat == 'Neutro' else '🔴')
    print(f'  {emoji} {cat}: {qtd:,} clientes ({pct:.1f}%)')
print(f'\n📉 Taxa de Detração: {df["is_detrator"].mean():.1%}')
```

```python
# Gráfico de Pizza — Composição do NPS
fig = go.Figure(data=[go.Pie(
    labels=contagem.index,
    values=contagem.values,
    hole=0.45,
    marker=dict(colors=[VERMELHO if c == 'Detrator' else AMARELO if c == 'Neutro' else VERDE
                         for c in contagem.index],
                line=dict(color='#0e1117', width=3)),
    textinfo='label+percent',
    textfont_size=14
)])
fig.update_layout(
    title=dict(text='🎯 Composição do NPS — Distribuição de Satisfação', font=dict(size=18)),
    paper_bgcolor='#1a1a2e', font=dict(color='white'),
    annotations=[dict(text=f'{df["is_detrator"].mean():.0%}<br>Detratores', x=0.5, y=0.5,
                      font=dict(size=16, color=VERMELHO), showarrow=False)]
)
fig.show()
```

## 📈 6. Curva de Risco × Dias de Atraso

Aqui verificamos **empiricamente** como a taxa de detração cresce conforme o atraso aumenta — confirmando o **Ponto de Ruptura nos 3 dias**.

```python
# Calcular taxa de detração por dias de atraso (agrupado)
curva = df.groupby('delivery_delay_days')['is_detrator'].agg(['mean', 'count']).reset_index()
curva.columns = ['delay_days', 'taxa_detracao', 'volume']
curva = curva[curva['delay_days'] <= 15]  # foco nos mais comuns
curva['taxa_detracao_pct'] = curva['taxa_detracao'] * 100

fig = make_subplots(specs=[[{'secondary_y': True}]])

fig.add_trace(go.Scatter(
    x=curva['delay_days'], y=curva['taxa_detracao_pct'],
    mode='lines+markers', name='Taxa de Detração (%)',
    line=dict(color=VERMELHO, width=3),
    marker=dict(size=8, color=VERMELHO)
), secondary_y=False)

fig.add_trace(go.Bar(
    x=curva['delay_days'], y=curva['volume'],
    name='Volume de Pedidos', opacity=0.25,
    marker_color=AZUL
), secondary_y=True)

fig.add_vline(x=3, line_dash='dash', line_color='orange', line_width=2)
fig.add_annotation(x=3, y=60, text='⚠️ Ruptura\n3 dias', showarrow=True,
                   arrowcolor='orange', font=dict(color='orange', size=12))

fig.update_layout(
    title='📈 Curva de Risco: Taxa de Detração × Dias de Atraso',
    paper_bgcolor='#1a1a2e', plot_bgcolor='#16213e',
    font=dict(color='white'), height=450,
    xaxis_title='Dias de Atraso', legend=dict(orientation='h')
)
fig.update_yaxes(title_text='Taxa de Detração (%)', ticksuffix='%', secondary_y=False)
fig.update_yaxes(title_text='Volume de Pedidos', secondary_y=True)
fig.show()

ruptura_pct = curva[curva['delay_days'] == 3]['taxa_detracao_pct'].values
if len(ruptura_pct) > 0:
    print(f'🚨 No 3º dia de atraso, a taxa de detração é de: {ruptura_pct[0]:.1f}%')
```

## 🔥 7. Heatmap de Correlação

Identificamos quais variáveis têm maior força de relação com o `nps_score`. Variáveis com alta correlação negativa são as mais perigosas para a satisfação.

```python
cols_corr = ['nps_score', 'delivery_delay_days', 'delivery_time_days',
             'customer_service_contacts', 'complaints_count', 'resolution_time_days',
             'order_value', 'items_quantity', 'csat_internal_score',
             'customer_tenure_months', 'delivery_attempts', 'payment_installments']

corr_matrix = df[cols_corr].corr()

fig, ax = plt.subplots(figsize=(13, 9))
fig.patch.set_facecolor('#1a1a2e')
ax.set_facecolor('#16213e')

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
            cmap='RdYlGn', center=0, square=True, ax=ax,
            linewidths=0.5, linecolor='#2d2d5e',
            cbar_kws={'shrink': 0.8})

ax.set_title('🔥 Matriz de Correlação — Identificação de Variáveis Preditoras',
             fontsize=14, pad=15, color='white')
plt.xticks(rotation=45, ha='right', color='white')
plt.yticks(color='white')
plt.tight_layout()
plt.show()

# Top correlações com NPS
corr_nps = corr_matrix['nps_score'].drop('nps_score').sort_values()
print('\n🏆 Top variáveis com MAIOR impacto negativo no NPS:')
for var, val in corr_nps.head(4).items():
    print(f'  🔴 {var}: {val:.3f}')
print('\n✅ Top variáveis com MAIOR impacto positivo no NPS:')
for var, val in corr_nps.tail(4).items():
    print(f'  🟢 {var}: {val:.3f}')
```

## 🌍 8. Análise Regional — Detração por Região

```python
regional = df.groupby('customer_region').agg(
    taxa_detracao=('is_detrator', 'mean'),
    atraso_medio=('delivery_delay_days', 'mean'),
    volume=('customer_id', 'count'),
    nps_medio=('nps_score', 'mean')
).reset_index().round(2)
regional['taxa_detracao_pct'] = (regional['taxa_detracao'] * 100).round(1)

fig = make_subplots(rows=1, cols=2, subplot_titles=('Taxa de Detração por Região (%)',
                                                     'Atraso Médio por Região (dias)'))

cores_reg = [VERMELHO if v == regional['taxa_detracao_pct'].max() else AZUL
             for v in regional['taxa_detracao_pct']]

fig.add_trace(go.Bar(x=regional['customer_region'], y=regional['taxa_detracao_pct'],
                     marker_color=cores_reg, text=regional['taxa_detracao_pct'],
                     texttemplate='%{text}%', textposition='outside', name='Detração'),
              row=1, col=1)

fig.add_trace(go.Bar(x=regional['customer_region'], y=regional['atraso_medio'],
                     marker_color=ROXO, text=regional['atraso_medio'],
                     texttemplate='%{text:.1f}d', textposition='outside', name='Atraso'),
              row=1, col=2)

fig.update_layout(title='🌍 Análise Regional de Risco',
                  paper_bgcolor='#1a1a2e', plot_bgcolor='#16213e',
                  font=dict(color='white'), height=420, showlegend=False)
fig.show()
print(regional.to_string(index=False))
```

## 🧪 9. Teste de Hipótese (Mann-Whitney U)

**CRISP-DM — Aula 3: Estatística Inferencial**

> **Hipótese nula (H₀):** Não existe diferença estatisticamente significativa entre o NPS de pedidos atrasados e no prazo.
> 
> **Hipótese alternativa (H₁):** O atraso na entrega causa uma queda significativa no NPS (p < 0.05).

```python
nps_no_prazo  = df[df['delivery_delay_days'] == 0]['nps_score']
nps_atrasados = df[df['delivery_delay_days'] >  0]['nps_score']

stat, p_valor = stats.mannwhitneyu(nps_no_prazo, nps_atrasados, alternative='greater')

print('='*55)
print('🧪 TESTE DE HIPÓTESE — Mann-Whitney U')
print('='*55)
print(f'  Pedidos no prazo  → n={len(nps_no_prazo):,}  | NPS médio: {nps_no_prazo.mean():.2f}')
print(f'  Pedidos atrasados → n={len(nps_atrasados):,} | NPS médio: {nps_atrasados.mean():.2f}')
print(f'  Estatística U: {stat:,.0f}')
print(f'  p-valor: {p_valor:.2e}')
print()
if p_valor < 0.05:
    print('  ✅ CONCLUSÃO: Rejeitamos H₀!')
    print('  O atraso causa queda significativa no NPS (p < 0.05).')
    print('  Nossa IA está embasada estatisticamente! 🚀')
else:
    print('  ❌ CONCLUSÃO: Não rejeitamos H₀.')
print('='*55)
```
